### This notebook presents several activation functions, included as part of the Neural Networks chapter of the thesis.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Apply LaTeX-style font settings 
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "mathtext.fontset": "stix",
    "font.size": 18,
    "axes.titlesize": 18,
    "axes.labelsize": 18,
    "xtick.labelsize": 17,
    "ytick.labelsize": 17,
    "legend.fontsize": 17,
    "figure.titlesize": 18,
})

# Activation functions
def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sign(x):
    return np.sign(x)

# Input values
x = np.linspace(-10, 10, 400)


y_relu = relu(x)
y_linear = linear(x)
y_sigmoid = sigmoid(x)
y_sign = sign(x)

# Plots 

# ReLU Activation Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_relu, color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
# plt.title("ReLU Activation Function")
plt.grid(True)
plt.show()

# Linear Activation Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_linear, color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
# plt.title("Linear Activation Function")
plt.grid(True)
plt.show()


# Sigmoid Activation Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_sigmoid, color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
# plt.title("Sigmoid Activation Function")
plt.grid(True)
plt.show()


# Sign Activation Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_sign, color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
# plt.title("Sign Activation Function")
plt.grid(True)
plt.show()
```
